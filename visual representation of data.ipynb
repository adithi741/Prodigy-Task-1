# Population of Specific Countries in "2004" :

```python
countries = ['South Africa','Zimbabwe','Aruba','Afghanistan','India','China','Brazil','Sri Lanka','Russia','Turkey','Japan']
filter_data = data[data['Country Name'].isin(countries)]
plt.figure(figsize=(15, 6))
plt.title("Population of Specific Countries in 2004")
sns.barplot(x='Country Name', y='2004', data=filter_data)
plt.ylabel("Population")
plt.xlabel("Countries")
plt.show()
```

# Distribution of 2004 Population with KDE by Country

```python
plt.figure(figsize=(15, 6))
sns.histplot(data=filter_data, x="2004", kde=True, hue="Country Name", element="step")

plt.title("Distribution of 2004 Population with KDE by Country")
plt.xlabel("Population in 2004")
plt.ylabel("Frequency")

plt.legend(title="Country")
plt.show()
```

# Distribution of Population with KDE by Country over the years.


```python
years = ['1960', '1986', '2004', '2014', '2021']

plt.figure(figsize=(15, 6))
for year in years:
    sns.histplot(data=filter_data, x=year, kde=True, hue="Country Name", element="step")

plt.title("Distribution of Population with KDE by Country Over the Years")
plt.xlabel("Population")
plt.ylabel("Frequency")

plt.legend(title="Country")
plt.show()
```

# Histogram.

```python
data['population_change'] = data['1965'] - data['2002']

# Histogram
plt.figure(figsize=(15,6))
plt.hist(data['population_change'], bins=10, color='blue', edgecolor='black')
plt.title('Distribution of Population Change (1965-2002)')
plt.xlabel('Population Change')
plt.ylabel('Number of Countries')
plt.grid(True, which="both", ls="--", linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
data['percentage_increase'] = ((data['2002'] - data['2020']) / data['2020']) * 100
top_countries = data.nlargest(5, 'percentage_increase')
```

# Population of Countries with Rapid Growth from 2015 to 2020 :

```python
bar_width = 0.35
index = range(len(top_countries))

plt.figure(figsize=(15, 8))
bar1 = plt.bar(index, top_countries['2015'], bar_width, label='2015', color='b', edgecolor='black')
bar2 = plt.bar([i+bar_width for i in index], top_countries['2015'], bar_width, label='2020', color='r', edgecolor='black')

plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population of Countries with Rapid Growth from 2015 to 2020')
plt.xticks([i+bar_width/2 for i in index], top_countries['Country Name'])
plt.legend()
plt.tight_layout()
plt.show()
```

# Population of Countries with Least Growth from 2000 to 2020

```python
data['percentage_increase'] = ((data['2000'] - data['2020']) / (data['2020'] + 1e-10)) * 100

bottom_countries = data.nsmallest(8, 'percentage_increase')

bar_width = 0.35
index = range(len(bottom_countries))

plt.figure(figsize=(15, 8))
bar1 = plt.bar(index, bottom_countries['2020'], bar_width, label='2020', color='b', edgecolor='black')
bar2 = plt.bar([i+bar_width for i in index], bottom_countries['2000'], bar_width, label='2000', color='r', edgecolor='black')

plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population of Countries with Least Growth from 2000 to 2020')
plt.xticks([i+bar_width/2 for i in index], bottom_countries['Country Name'])
plt.legend()
plt.tight_layout()
plt.show()
```

```python
data['population_change_2015_2021'] = data['2015'] - data['2021']

country_with_largest_dip = data.nsmallest(5, 'population_change_2015_2021')

print(country_with_largest_dip[['Country Name', 'population_change_2015_2021']])
```

# Population of India vs China from 1960 to 2021 :

```python
country_data_1 = data[data['Country Name'] == 'India'].iloc[0]
country_data_2 = data[data['Country Name'] == 'China'].iloc[0]

years = ['1960', '1965', '1975', '1980', '2000', '2002', '2010','2015','2021']
population_values_1 = country_data_1[years]
population_values_2 = country_data_2[years]

bar_width = 0.35
index = range(len(years))

plt.figure(figsize=(15, 8))
bars1 = plt.bar(index, population_values_1, bar_width, label='India', color='blue', edgecolor='black')
bars2 = plt.bar([i+bar_width for i in index], population_values_2, bar_width, label='China', color='red', edgecolor='black')

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of India vs China from 1960 to 2021')
plt.xticks([i+bar_width/2 for i in index], years)
plt.legend()
plt.tight_layout()
plt.show()
```

# Population Trend from 1960 to 2021:

```python
plt.figure(figsize=(15, 8))
plt.plot(years, population_values_1, marker='o', label='India', color='blue')
plt.plot(years, population_values_2, marker='o', label='China', color='red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Trend from 1960 to 2021')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
